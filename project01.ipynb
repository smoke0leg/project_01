```python
# Formating education column
hh_df['Образование и ВУЗ'] = hh_df['Образование и ВУЗ'].apply(lambda x: x.split(' ')[0:2])
hh_df['Образование и ВУЗ'] = hh_df['Образование и ВУЗ'].apply(lambda x: x[0:1] if 'образование' in x else x)
hh_df['Образование и ВУЗ'] = hh_df['Образование и ВУЗ'].apply(lambda x: (x[0] + ' ' + x[1]) if len(x) == 2 else x[0])
hh_df['Образование и ВУЗ'] = hh_df['Образование и ВУЗ'].str.lower()
hh_df = hh_df.rename(columns={'Образование и ВУЗ': 'Образование'})
# Displaying data for answering on questions
print(hh_df[hh_df['Образование'] == 'среднее'].shape[0])
```

```python
# Creating two separate columns age and gender
hh_df['Возраст'] = hh_df['Пол, возраст'].apply(lambda x: int(x.split(',')[1][2:4]))
hh_df['Пол'] = hh_df['Пол, возраст'].apply(lambda x: 'М' if 'Мужчина' in x.split(',')[0] else 'Ж')
hh_df = hh_df.drop('Пол, возраст', axis=1)
# Displaying data for answering on questions
print(round(hh_df[hh_df['Пол'] == 'Ж'].shape[0]/hh_df.shape[0]*100, 2))
print(round(hh_df['Возраст'].mean(), 1))
```

```python
# Create new column job expirence in month and deleting job experience column
def get_month_experience(experience):
    """Function for converting string with job experience in month and years
    with job description, to int job experience in month

    Args:
        experience (string): job experience  with job description

    Returns:
        int: job experience in month
    """
    if type(experience) != str:
        return np.nan
    elif experience == 'Не указано':
        return np.nan
    else:
        experience = experience.split('  ')[0]
        experience = experience.split(' ')
        year_key_words = ['год', 'года', 'лет']
        month_key_words = ['месяц', 'месяца', 'месяцев']
        month = 0
        for word in experience:
            if word in year_key_words:
                month += (int(experience[(experience.index(word) - 1)])) * 12
            if word in month_key_words:
                month += int(experience[(experience.index(word) - 1)])
            
        return month
    
# Adding column job experience in month and delete column job experience with description
hh_df['Опыт работы (месяц)'] = hh_df['Опыт работы'].apply(get_month_experience)
hh_df = hh_df.drop('Опыт работы', axis=1)
# Displaying data for answering on questions
print(hh_df['Опыт работы (месяц)'].median())
```

```python
# Create three separate columns city, ready for business trips and ready for moving
def get_city(city_data):
    """Function for get city type from column (city, ready for move, 
    ready for business trips)

    Args:
        city_data (string): data about city, ready for move, ready for business trips

    Returns:
        string: city type 
    """
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород',
                      'Казань', 'Челябинск','Омск', 'Самара',
                      'Ростов-на-Дону', 'Уфа', 'Красноярск',
                      'Пермь', 'Воронеж','Волгоград']
    city = city_data.split(',')[0]
    city = city[:-1]
    if city in million_cities:
        city_type = 'город-миллионник'
    elif city == 'Москва':
        city_type = 'Москва'
    elif city == 'Санкт-Петербург':
        city_type = 'Санкт-Петербург'
    else:
        city_type = 'другие'
    return city_type


def get_ready_for_bisiness_trips(arg):
    """Function for get boolvalue of ready for business trips 
    from column (city, ready for move, ready for business trips)

    Args:
        arg (string): data about city, ready for move, ready for business trips

    Returns:
        bool: Ready for business trips
    """
    if ('командировка' in arg):
        if ('не готов к командировкам' in arg) or('не готова к командировкам' in arg):
            return False
        else: 
            return True
    else:
        return False
    
# Create new column with city type by applying function 'get_city'
hh_df['Город'] = hh_df['Город, переезд, командировки'].apply(get_city)
# Create new column with boolean ready for bussines trips by applying function 'get_ready_for_bisiness_trips'    
hh_df['Готовность к командировкам'] = hh_df['Город, переезд, командировки'].apply(get_ready_for_bisiness_trips)
# Create new column with boolean ready for moving
hh_df['Готовность к переезду'] = hh_df['Город, переезд, командировки'].apply(lambda x: False 
                                                                                 if 'не готов к переезду' in x or 
                                                                                 'не готова к переезду' in x else True)
# Delete column (city, ready for move, ready for business trips) after convertion
hh_df = hh_df.drop('Город, переезд, командировки', axis=1)
# Displaying data for answering on questions
print(round(hh_df[hh_df['Город'] == 'Санкт-Петербург'].shape[0]/hh_df.shape[0] * 100))
print(round(hh_df[(hh_df['Готовность к командировкам'] == True) & (hh_df['Готовность к переезду'] == True)].shape[0]/hh_df.shape[0]*100))
```

```python
# Creating new columns for each type of eployment
employments = ['полная занятость', 'частичная занятость',
              'проектная работа', 'волонтерство', 'стажировка']
for employment in employments:
    hh_df[employment] = hh_df['Занятость'].apply(lambda x: employment in x)
    
schedules = ['полный день', 'сменный график', 
         'гибкий график', 'удаленная работа',
         'вахтовый метод']

for schedule in schedules:
    hh_df[schedule] = hh_df['График'].apply(lambda x: schedule in x)
# Delete columns with schedule and employment after taking columns with types    
hh_df = hh_df.drop('Занятость', axis=1)
hh_df = hh_df.drop('График', axis=1)
# Displaying data for answering on questions
print(hh_df[(hh_df['проектная работа'] == True) & (hh_df['волонтерство'] == True)].shape[0])
print(hh_df[(hh_df['вахтовый метод'] == True) & (hh_df['гибкий график'])].shape[0])
```

```python
# Reading exchange rate dataframe
exchange_rates_df = pd.read_csv('data/ExchangeRates.csv')
exchange_rates_df = exchange_rates_df.drop('vol', axis=1)
exchange_rates_df = exchange_rates_df.drop('per', axis=1)
exchange_rates_df = exchange_rates_df.drop('time', axis=1)
exchange_rates_df['date'] = pd.to_datetime(exchange_rates_df['date'], dayfirst=True).dt.date
# Convert resume update column to datetime format
hh_df['Обновление резюме'] = pd.to_datetime(hh_df['Обновление резюме'], dayfirst=True).dt.date
# Convert salary column to int and takeing coin ticker
hh_df['ЗП'] = hh_df['ЗП'].apply(lambda x: x.split(' '))
hh_df['ЗП_валюта'] = hh_df['ЗП'].apply(lambda x: x[1])
hh_df['ЗП'] = (hh_df['ЗП'].apply(lambda x: x[0])).astype('float64')


def get_currency_iso(currency):
    """Function for convert currency ticker to iso format
    Args:
        currency (string): currency ticker 

    Returns:
        string: currency ticker in iso format
    """
    # currency = currency.split(' ')
    if currency == 'грн.':
        currency = 'UAH'
    if currency == 'бел.руб.':
        currency = 'BYN'
    if currency == 'сум':
        currency = 'UZS'
    if currency == ('руб.'):
        currency = 'RUB'
    return currency

# Applying function for get ISO format of coin ticker
hh_df['ЗП_валюта'] = hh_df['ЗП_валюта'].apply(get_currency_iso)
# Joining main dataframe with dataframe with exchange rates
hh_df = pd.merge(left=hh_df, right=exchange_rates_df, left_on=['Обновление резюме', 'ЗП_валюта'], right_on=['date', 'currency'], how='left')
# Data preparation before  converting salarys to RUB
hh_df['currency'] = hh_df['currency'].fillna('RUB')
hh_df['proportion'] = hh_df['proportion'].fillna(1)
hh_df['close'] = hh_df['close'].fillna(1)
# Converting salary to RUB and clean from exchange data
hh_df['ЗП руб'] = hh_df['ЗП'] * hh_df['close'] / hh_df['proportion']
hh_df = hh_df.drop(['proportion', 'close', 'date', 'currency', 'ЗП', 'ЗП_валюта'], axis=1)
# Displaying data for answering on questions
print(round(hh_df['ЗП руб'].median() / 1000))
```
